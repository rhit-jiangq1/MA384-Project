```python
soccer['str_positions'] = soccer['str_positions'].str.split(',')
soccer['dt_date_of_birth'] = pd.to_datetime(soccer['dt_date_of_birth'])
soccer['str_nationality'] = soccer['str_nationality'].astype('category')
soccer['str_best_position'] = soccer['str_best_position'].astype('category')
soccer['str_preferred_foot'] = soccer['str_preferred_foot'].astype('category')
soccer['str_work_rate'] = soccer['str_work_rate'].astype('category')
soccer['str_body_type'] = soccer['str_body_type'].astype('category')
soccer.head(3)
```
